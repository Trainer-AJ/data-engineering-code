# Mounting ADLS to Databricks
[Link](https://techcommunity.microsoft.com/t5/azure-paas-blog/mount-adls-gen2-or-blob-storage-in-azure-databricks/ba-p/3802926)

```
dbutils.widgets.text("Str","","Storage Account Name")
dbutils.widgets.text("key","","Storage Account Access key")
dbutils.widgets.text("blob","","Blob Container Name")

str = dbutils.widgets.get("Str")
# print(str)
blob = dbutils.widgets.get("key")
key = dbutils.widgets.get("blob")
```

```
print(str)
```

```
# spark.conf.set(
#     "fs.azure.account.key.<storage-account>.dfs.core.windows.net",
#     dbutils.secrets.get(scope="<scope>", key="<storage-account-access-key>"))

storageAccountName = str
storageAccountAccessKey = key
#sasToken = <sas-token>
blobContainerName = blob
mountPoint = "/mnt/data/"
if not any(mount.mountPoint == mountPoint for mount in dbutils.fs.mounts()):
  try:
    dbutils.fs.mount(
      source = f"abfs://{blobContainerName}@{storageAccountName}.dfs.core.windows.net/",
      mount_point = mountPoint,
      extra_configs = {'fs.azure.account.key.' + storageAccountName + '.blob.core.windows.net': storageAccountAccessKey}
      #extra_configs = {'fs.azure.sas.' + blobContainerName + '.' + storageAccountName + '.blob.core.windows.net': sasToken}
    )
    print("mount succeeded!")
  except Exception as e:
    print("mount exception", e)
```
